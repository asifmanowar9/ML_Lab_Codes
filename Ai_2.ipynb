```python
import pandas as pd
import numpy as np
```

```python
dataf = pd.read_csv('/content/bd_students_per_v2.csv')
display(dataf.head())
```

```python
dataf.isnull().sum()
```

```python
dataf = dataf.dropna()
dataf.isnull().sum()
```

```python
# There is no duplicated values
dataf.duplicated().sum()
```

```python
# We take two label
# 1. pass_fail which predict the students whether passed or failed
# 2. cgpa will predict the the students grade
# Here we also take take new column 'avg_score' of english and math and function
# cf() just for an example.

dataf["avg_score"] = (dataf["english"]+dataf["math"])/2
dataf["pass_fail"] = (dataf["avg_score"] >=60).astype(int)
def cg(score):
    if score >= 80:
      return 4.0
    elif score >= 70:
      return 3.5
    elif score >= 60:
      return 3.0
    else:
      return 2.5
dataf["cgpa"] = dataf["avg_score"].apply(cg)
display(dataf[["full_name", "english", "math", "avg_score", "pass_fail", "cgpa"]])
```
