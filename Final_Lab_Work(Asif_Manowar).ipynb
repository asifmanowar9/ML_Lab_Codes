**from the boxplots we can see that bmi, s1, s2, s3, s4, s5 and s6 have outliers**

example of outliers with the help of IQR(quantile) method

```python
Q1_bmi = df_diabetes['bmi'].quantile(0.25)
Q3_bmi = df_diabetes['bmi'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi

upper_bound_bmi = Q3_bmi + 1.5 * IQR_bmi
lower_bound_bmi = Q1_bmi - 1.5 * IQR_bmi

print(f"Q1 (bmi): {Q1_bmi}")
print(f"Q3 (bmi): {Q3_bmi}")
print(f"IQR (bmi): {IQR_bmi}")
print(f"Upper Bound (bmi): {upper_bound_bmi}")
print(f"Lower Bound (bmi): {lower_bound_bmi}")

outliers_bmi = df_diabetes[(df_diabetes['bmi'] < lower_bound_bmi) | (df_diabetes['bmi'] > upper_bound_bmi)]

print("\nOutliers in 'bmi' column:")
if not outliers_bmi.empty:
    display(outliers_bmi[['bmi']])
else:
    print("No outliers found in 'bmi' column.")
```

**from the dataset i think all the data has already been scalled or normalized. so scalling and normalization is not need from my side, but i did apply standard scaler because without it Linear regression performend poorly like the r2 score was in the negative(-ve) but after the scaling it has gone to 0.32**

```python
from sklearn.preprocessing import StandardScaler

feature_columns = df_diabetes.drop(columns=['target']).columns

scaler = StandardScaler()

df_diabetes[feature_columns] = scaler.fit_transform(df_diabetes[feature_columns])


df_diabetes.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_diabetes.corr()

print("Correlation Matrix:")
display(correlation_matrix)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
import numpy as np

# Set a threshold for high correlation
threshold = 0.7

# Find highly correlated pairs
highly_correlated_pairs = []
# Use a mask to avoid duplicate pairs and self-correlation
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        col1 = correlation_matrix.columns[i]
        col2 = correlation_matrix.columns[j]
        correlation_value = correlation_matrix.loc[col1, col2]
        if abs(correlation_value) > threshold:
            highly_correlated_pairs.append((col1, col2, correlation_value))

if highly_correlated_pairs:
    print(f"Features with an absolute correlation greater than {threshold}:")
    for pair in highly_correlated_pairs:
        print(f"- {pair[0]} and {pair[1]}: {pair[2]:.2f}")
else:
    print(f"No features found with an absolute correlation greater than {threshold}.")
```

```python
from sklearn.model_selection import train_test_split


X = df_diabetes.drop('target', axis=1)
y = df_diabetes['target']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report

model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)


print("accuracy: " , accuracy_score(y_test, y_pred))
print("\nclassification report: ", classification_report(y_test, y_pred))

```

**logistic regression is a classification model, so it is not suitble for this dataset**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score, root_mean_squared_error

model_lr = LinearRegression()

model_lr.fit(X_train, y_train)

y_pred_lr = model.predict(X_test)


print("mean absolute error: " , mean_absolute_error(y_test, y_pred_lr))
print("\nmean squared error: ", mean_squared_error(y_test, y_pred_lr))
print("\nr2 score: ", r2_score(y_test, y_pred_lr))
print("\nroot mean squared error: ", root_mean_squared_error(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(random_state=42)

model_rf.fit(X_train, y_train)

y_pred_rf = model_rf.predict(X_test)

print("mean absolute error: " , mean_absolute_error(y_test, y_pred_rf))
print("\nmean squared error: ", mean_squared_error(y_test, y_pred_rf))
print("\nr2 score: ", r2_score(y_test, y_pred_rf))
print("\nroot mean squared error: ", root_mean_squared_error(y_test, y_pred_rf))
```

```python
from sklearn.svm import SVR

model_svr = SVR()

model_svr.fit(X_train, y_train)

y_pred_svr = model_svr.predict(X_test)

print("mean absolute error: " , mean_absolute_error(y_test, y_pred_svr))
print("\nmean squared error: ", mean_squared_error(y_test, y_pred_svr))
print("\nr2 score: ", r2_score(y_test, y_pred_svr))
print("\nroot mean squared error: ", root_mean_squared_error(y_test, y_pred_svr))
```

```python
from xgboost import XGBRegressor

model_xgb = XGBRegressor(random_state=42)
model_xgb.fit(X_train, y_train)

y_pred_xgb = model_xgb.predict(X_test)

print("mean absolute error: " , mean_absolute_error(y_test, y_pred_xgb))
print("\nmean squared error: ", mean_squared_error(y_test, y_pred_xgb))
print("\nr2 score: ", r2_score(y_test, y_pred_xgb))
print("\nroot mean squared error: ", root_mean_squared_error(y_test, y_pred_xgb))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# 2. Define a dictionary param_grid specifying the hyperparameters to tune
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': [0.6, 0.8, 1.0],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# 3. Initialize GridSearchCV
# Note: max_features='auto' is deprecated in scikit-learn 1.1 and removed in 1.3, 'sqrt' is the default.
# Using numerical values (e.g., 0.6, 0.8, 1.0) or 'sqrt' explicitly.
# 'neg_mean_squared_error' is a common scoring metric for regression tasks in GridSearchCV.
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1, # Use all available processors
    verbose=2
)

# 4. Fit the grid_search object to the training data
print("Fitting GridSearchCV...")
grid_search.fit(X_train, y_train)
print("GridSearchCV fitting complete.")

# 5. Print the best parameters found by GridSearchCV
print("\nBest parameters found:", grid_search.best_params_)

# 6. Store the best estimator
best_rf_model = grid_search.best_estimator_
print("Best Random Forest model stored in 'best_rf_model'.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Make predictions on the test set using the best model
y_pred_tuned_rf = best_rf_model.predict(X_test)

# Evaluate the tuned model
mae_tuned_rf = mean_absolute_error(y_test, y_pred_tuned_rf)
mse_tuned_rf = mean_squared_error(y_test, y_pred_tuned_rf)
rmse_tuned_rf = np.sqrt(mse_tuned_rf)
r2_tuned_rf = r2_score(y_test, y_pred_tuned_rf)

print("\n--- Tuned Random Forest Regressor Performance ---")
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Mean Absolute Error (MAE): {mae_tuned_rf:.4f}")
print(f"Mean Squared Error (MSE): {mse_tuned_rf:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_tuned_rf:.4f}")
print(f"R2 Score: {r2_tuned_rf:.4f}")

# For comparison, print the untuned model's performance again
print("\n--- Untuned Random Forest Regressor Performance ---")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred_rf):.4f}")
print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred_rf):.4f}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test, y_pred_rf)):.4f}")
print(f"R2 Score: {r2_score(y_test, y_pred_rf):.4f}")
```
